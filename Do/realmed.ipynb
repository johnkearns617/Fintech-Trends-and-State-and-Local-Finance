```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time

from fredapi import Fred
fred_key = '170531604ba9e3b805b9635fbd6aa9b7'
```

```python
fred = Fred(api_key=fred_key)
```

```python
realmed_df = fred.search('real median income', filter=('frequency', 'Annual'))
realmed_df = realmed_df.loc[realmed_df['title'].str.contains('Real Median Household Income in')]
print(realmed_df.shape)
```

```python
all_results = []

for myid in realmed_df.index:
    results = fred.get_series(myid)
    results = results.to_frame(name=myid)
    all_results.append(results)
    time.sleep(0.1)
realmed_results = pd.concat(all_results, axis=1)
```

```python
print(realmed_results.head)
```

```python
id_to_state = realmed_df['title'].str.replace('Real Median Household Income in ','').to_dict()
realmed_results.columns = [id_to_state[c] for c in realmed_results.columns]
```

```python
print(realmed_results.head)
```

```python
df = realmed_results
df.to_csv('FREDdata5.csv')
```
